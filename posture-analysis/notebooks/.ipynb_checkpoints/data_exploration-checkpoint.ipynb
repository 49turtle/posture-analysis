```python
import json
import matplotlib.pyplot as plt
```

```python
# AI-Hub JSON 파일 로드 함수
def load_ai_hub_json(path):
    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    keypoints = []
    for annotation in data['annotations']:
        keypoints.append(annotation['keypoints'])
    return keypoints

# OpenPose JSON 파일 로드 함수
def load_openpose_json(path):
    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    keypoints = []
    for frame in data:
        for person in frame['keypoints']:
            keypoints.append(person)
    return keypoints
```

```python
# 시각화 함수 정의
def visualize_ai_hub_keypoints(keypoints, labels, title="Keypoints"):
    for person in keypoints[:1]:  # 첫 번째 사람만 시각화
        x = person[0::3]
        y = person[1::3]
        # (0, 0) 좌표 필터링
        x_filtered = [x_i for x_i, y_i in zip(x, y) if not (x_i == 0 and y_i == 0)]
        y_filtered = [y_i for x_i, y_i in zip(x, y) if not (x_i == 0 and y_i == 0)]
        labels_filtered = [label for x_i, y_i, label in zip(x, y, labels) if not (x_i == 0 and y_i == 0)]
        
        plt.scatter(x_filtered, y_filtered)
        for i, label in enumerate(labels_filtered):
            plt.text(x_filtered[i], y_filtered[i], label, fontsize=8, ha='right')
    plt.gca().invert_yaxis()  # 이미지 좌표계를 고려하여 y축을 반전
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

# OpenPose 시각화 함수 정의
def visualize_openpose_keypoints(keypoints, labels, title="OpenPose Keypoints"):
    for person in keypoints[:1]:  # 첫 번째 사람만 시각화
        x = [point[0] for point in person if not (point[0] == 0 and point[1] == 0)]
        y = [point[1] for point in person if not (point[0] == 0 and point[1] == 0)]
        labels_filtered = [label for point, label in zip(person, labels) if not (point[0] == 0 and point[1] == 0)]
        
        plt.scatter(x, y)
        for i, label in enumerate(labels_filtered):
            plt.text(x[i], y[i], label, fontsize=8, ha='right')
    plt.gca().invert_yaxis()  # 이미지 좌표계를 고려하여 y축을 반전
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
# AI-Hub 시각화 함수 정의
def visualize_ai_hub_keypoints_with_lines(keypoints, labels, connections, title="Keypoints with Connections"):
    for person in keypoints[:1]:  # 첫 번째 사람만 시각화
        x = person[0::3]
        y = person[1::3]
        # (0, 0) 좌표 필터링
        x_filtered = [x_i for x_i, y_i in zip(x, y) if not (x_i == 0 and y_i == 0)]
        y_filtered = [y_i for x_i, y_i in zip(x, y) if not (x_i == 0 and y_i == 0)]
        labels_filtered = [label for x_i, y_i, label in zip(x, y, labels) if not (x_i == 0 and y_i == 0)]
        index_filtered = [i for i, (x_i, y_i) in enumerate(zip(x, y)) if not (x_i == 0 and y_i == 0)]
        
        plt.scatter(x_filtered, y_filtered)
        for i, label in enumerate(labels_filtered):
            plt.text(x_filtered[i], y_filtered[i], label, fontsize=8, ha='right')
        
        for (i, j) in connections:
            if i in index_filtered and j in index_filtered:
                xi, yi = x[i], y[i]
                xj, yj = x[j], y[j]
                plt.plot([xi, xj], [yi, yj], 'r-')
        
    plt.gca().invert_yaxis()  # 이미지 좌표계를 고려하여 y축을 반전
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

# OpenPose 시각화 함수 정의
def visualize_openpose_keypoints_with_lines(keypoints, labels, connections, title="OpenPose Keypoints with Connections"):
    for person in keypoints[:1]:  # 첫 번째 사람만 시각화
        x = [point[0] for point in person if not (point[0] == 0 and point[1] == 0)]
        y = [point[1] for point in person if not (point[0] == 0 and point[1] == 0)]
        labels_filtered = [label for point, label in zip(person, labels) if not (point[0] == 0 and point[1] == 0)]
        index_filtered = [i for i, point in enumerate(person) if not (point[0] == 0 and point[1] == 0)]
        
        plt.scatter(x, y)
        for i, label in enumerate(labels_filtered):
            plt.text(x[i], y[i], label, fontsize=8, ha='right')
        
        for (i, j) in connections:
            if i in index_filtered and j in index_filtered:
                xi, yi = person[i][0], person[i][1]
                xj, yj = person[j][0], person[j][1]
                plt.plot([xi, xj], [yi, yj], 'r-')
        
    plt.gca().invert_yaxis()  # 이미지 좌표계를 고려하여 y축을 반전
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# keypoints
neck = [2, 4]
head_top = [2, 6]
neck_vertical = [neck[0], neck[1] - 1]

def calculate_angle(p1, p2, p3):
    a = np.array(p1)
    b = np.array(p2)
    c = np.array(p3)
    
    ba = a - b
    bc = c - b
    
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(cosine_angle)
    
    return np.degrees(angle)

# Calculate the angle
angle = calculate_angle(neck, head_top, neck_vertical)

# Plot points
plt.scatter(*zip(neck, head_top, neck_vertical), color=['red', 'blue', 'green'])
plt.text(neck[0], neck[1], 'neck', fontsize=12, ha='right')
plt.text(head_top[0], head_top[1], 'head_top', fontsize=12, ha='right')
plt.text(neck_vertical[0], neck_vertical[1], 'neck_vertical', fontsize=12, ha='right')

# Plot lines
plt.plot([neck[0], head_top[0]], [neck[1], head_top[1]], color='blue')
plt.plot([neck[0], neck_vertical[0]], [neck[1], neck_vertical[1]], color='green')

plt.xlim(0, 4)
plt.ylim(0, 8)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Neck and Head Angle Visualization\nAngle: {angle:.2f} degrees')
plt.show()
```
